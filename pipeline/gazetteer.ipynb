```python
import os
import requests
import zipfile
import tempfile
import pandas as pd
```

```python
import glob

fields = '''
geonameid
name
asciiname
alternatenames
latitude
longitude
feature class
feature code
country code
cc2
admin1 code
admin2 code
admin3 code
admin4 code
population
elevation
dem
timezone
modification date
'''.strip().split('\n')
fields = [f.strip() for f in fields]
```

```python
local_zip = 'local.zip'
dataset = 'GB'
with tempfile.TemporaryDirectory() as tmp:
    current_dir = os.getcwd()
    try:
        os.chdir(tmp)
        r = requests.get(
            f'http://download.geonames.org/export/dump/{dataset}.zip')
        with open(local_zip, 'wb') as f:
            f.write(r.content)

        with zipfile.ZipFile(local_zip, 'r') as zip:
            zip.extractall('.')
        print(glob.glob('*'))
        data = pd.read_csv(f'{dataset}.txt', sep='\t',
                           header=None, names=fields)
    finally:
        os.chdir(current_dir)
```
